```python
import random
import matplotlib.pyplot as plt
```

```python
points = 100
minx = -5
maxx = 5
miny = -5
maxy = 5
f = plt.figure()
f.set_figwidth(5)
f.set_figheight(5)
```

```python
perfect_agreement_x = []
perfect_agreement_y = []
for _ in range(points):
    x = random.random()*10-5
    y = x
    perfect_agreement_x.append(x)
    perfect_agreement_y.append(y)
```

```python
plt.scatter(perfect_agreement_x, perfect_agreement_y, s=10)
plt.title('Perfect Agreement')
plt.plot([-5,5],[-5.05,4.95],color = 'r')
plt.plot([-5,5],[-4.95,5.05],color = 'orange')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['Perfect Linear Agreement','Perfect Agreement', 'Perfect Angular Agreement'])
plt.show()
```

```python
bad_linear_agreement_x = []
bad_linear_agreement_y = []
for _ in range(points):
    x = random.random()*10-5
    y = x+random.random()*8-4
    bad_linear_agreement_x.append(x)
    bad_linear_agreement_y.append(y)
```

```python
plt.scatter(bad_linear_agreement_x, bad_linear_agreement_y, s=10)
plt.title('Bad Linear Agreement')
plt.plot([-5,5],[-5.05,4.95],color = 'r')
plt.plot([-5,5],[-4.95,5.05],color = 'orange')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['Bad Linear Agreement','Perfect Agreement', 'Perfect Angular Agreement'])
plt.show()
```

```python
bad_angular_agreement_x = []
bad_angular_agreement_y = []
for _ in range(points):
    x = random.random()*2-1
    y = -1*x+random.random()*0.25-0.125
    bad_angular_agreement_x.append(x)
    bad_angular_agreement_y.append(y)
```

```python
plt.scatter(bad_angular_agreement_x, bad_angular_agreement_y, s=10)
plt.title('Bad Angular Agreement')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot([-5,5],[-5,5],color = 'r')
plt.plot([2,-2],[-2,2],color = 'orange')
plt.legend(['Decent Linear Agreement','Perfect Agreement', 'Bad Angular Agreement'])
plt.show()
```

```python
disagreement_x = []
disagreement_y = []
for _ in range(points):
    x = random.random()*10-5
    y = random.random()*10-5
    disagreement_x.append(x)
    disagreement_y.append(y)
```

```python
plt.scatter(disagreement_x, disagreement_y, s=10)
plt.title('Total Disagreement')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot([-5,5],[-5,5],color = 'r')
plt.plot([-5,5],[0, 0],color = 'orange')
plt.legend(['Bad Linear Agreement','Perfect Agreement', 'Bad Angular Agreement'])
plt.show()
```
